```python
print(df['Genre'].value_counts())
plt.figure(figsize=(10,8))
sns.countplot(x=df['Genre'])
plt.xticks(rotation=90)
plt.show()
```

```python
labels = ['Documentary', 'Thriller', 'Crime', 'Reality', 'Rom-Com', 'Horror', 'Family', 'Drama', 'Comedy', 'Adventure', 'Sci-Fi', 'Cartoon', 'Action', 'Anime', 'LGBTQ', 'Sports', 'Music']
size=[1850, 1542, 1048, 838, 625, 539, 515, 425, 422, 305, 290, 247, 67, 37, 26, 17, 10]
plt.pie(size, labels=labels, autopct='%0.1f%%')
plt.title('Pie Chart', c='r')
plt.show()
```

```python
fig,axs=plt.subplots(2,1,figsize=(12,7))
sns.countplot(x=df["type"],data=df,ax=axs[0])
sns.countplot(x=df["Genre"],data=df,ax=axs[1])
```

```python
genre = ['Documentary', 'Thriller', 'Crime', 'Reality', 'Rom-Com', 'Horror', 'Family', 'Drama', 'Comedy', 'Adventure', 'Sci-Fi', 'Cartoon', 'Action', 'Anime', 'LGBTQ', 'Sports', 'Music']
size=np.array([1850, 1542, 1048, 838, 625, 539, 515, 425, 422, 305, 290, 247, 67, 37, 26, 17, 10])
plt.scatter(genre, size)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Scatter Plot',c='r')
plt.show()
```
